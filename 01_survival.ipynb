```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error

import xgboost as xgb
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
```



---



```python
def __indiviual_runs__(df, df_name, df_survival, regressor, train_size, random_state):
    X_train, X_test, y_train, y_test = train_test_split(df, df_survival, train_size = train_size, random_state = random_state)

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    scaler_os = StandardScaler()
    y_train_scaled = scaler_os.fit_transform(y_train)
    y_test_scaled = scaler_os.transform(y_test)

    # regressor = KNeighborsRegressor()
    regressor.fit(X_train_scaled, y_train_scaled)
    y_pred = regressor.predict(X_test_scaled)
    ## mean squared error
    mse = mean_squared_error(y_test_scaled, y_pred)

    print(type(regressor).__name__)
    print("Mean Squared Error:", mse)
    ## mean squared error
    mae = mean_absolute_error(y_test_scaled, y_pred)
    print("Mean Absolute Error:", mae)
    ## mean absolute percentage error
    mape = mean_absolute_percentage_error(y_test_scaled, y_pred)
    print("Mean Absolute Percentage Error:", mape)
    print()

    dict = {'mse': mse, 'mae': mae, 'mape': mape}

    with open(OUTPUT_PATH+f"{type(regressor).__name__}_seed{random_state}_{df_name}.txt", 'w') as f:
        f.write(repr(dict) + '\n')
```

```python
'''
CHANGE THE VALUE OF random_state AFTER EACH RUN. KEEP IT BETWEEN 0 to 65535.
'''

train_size = 0.8
random_state = 100
regressor_knn = KNeighborsRegressor()
regressor_svr = SVR()
regressor_xgb = xgb.XGBRegressor(objective = 'reg:squarederror', random_state = random_state)
```

INDIVIDUAL RUN ON DF_BRCA

```python
__indiviual_runs__(df_brca_imputed, 'df_brca', df_survival_imputed_brca, regressor_knn, train_size, random_state)
__indiviual_runs__(df_brca_imputed, 'df_brca', df_survival_imputed_brca, regressor_svr, train_size, random_state)
__indiviual_runs__(df_brca_imputed, 'df_brca', df_survival_imputed_brca, regressor_xgb, train_size, random_state)
```

INDIVIDUAL RUN FOR LUNG

```python
__indiviual_runs__(df_lung_imputed, 'df_lung', df_survival_imputed_lung, regressor_knn, train_size, random_state)
__indiviual_runs__(df_lung_imputed, 'df_lung', df_survival_imputed_lung, regressor_svr, train_size, random_state)
__indiviual_runs__(df_lung_imputed, 'df_lung', df_survival_imputed_lung, regressor_xgb, train_size, random_state)
```

INDIVIDUAL RUN FOR COAD

```python
__indiviual_runs__(df_coad_imputed, 'df_coad', df_survival_imputed_coad, regressor_knn, train_size, random_state)
__indiviual_runs__(df_coad_imputed, 'df_coad', df_survival_imputed_coad, regressor_svr, train_size, random_state)
__indiviual_runs__(df_coad_imputed, 'df_coad', df_survival_imputed_coad, regressor_xgb, train_size, random_state)
```

INDIVIDUAL RUN FOR COMBINED DATASET

```python
__indiviual_runs__(df_concat, 'df_concat', survival_concat, regressor_knn, train_size, random_state)
__indiviual_runs__(df_concat, 'df_concat', survival_concat, regressor_svr, train_size, random_state)
__indiviual_runs__(df_concat, 'df_concat', survival_concat, regressor_xgb, train_size, random_state)
```
